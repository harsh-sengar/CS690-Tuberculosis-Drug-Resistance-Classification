```
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['File_Name.csv'])) #Enter the file name here.
# Dataset is now stored in a Pandas Dataframe
```

```
#Replacing -1 with Nan

import numpy as np
new_df = df.replace(-1,np.nan)
new_df
```

```
missing_col = new_df.columns[new_df.isnull().any()].tolist() #Print the columns having missing values.
```

```
#Taking only those columns as a new dataframe having missing values
df2 = new_df[missing_col]
df2
```

```
# Count total NaN at each column in a DataFrame and calculate the percentage of missing values in the three columns
print(" \nCount total NaN at each column in a DataFrame : \n\n", 
      (df2.isnull().sum()*100/len(df.index))) 
```

```
#Printing the number of missing values in each of the row in our original dataset (containing all columns)

for i in range(len(new_df.index)) : 
    print(" Total NaN in row", i + 1, ":", 
          new_df.iloc[i].isnull().sum()) 
    
```
