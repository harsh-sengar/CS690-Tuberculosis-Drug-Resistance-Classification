```
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['X_trainData_column_modified_EMB.csv'])) #Enter the file name here.
# Dataset is now stored in a Pandas Dataframe
```

```
df
```

```
#Dividing the dataset for oversampling
X = df.iloc[:,:-1] 
Y = df.iloc[:,-1]
```

```
from imblearn.over_sampling import RandomOverSampler # Importing the RandomOverSampler class
```

```
oversample = RandomOverSampler(sampling_strategy=1) # This floating value decides the new ratio of minority to majority (minority/majority)
```

```
X_sample, Y_sample = oversample.fit_resample(X,Y) # Y contains the data on which sampling is done (target array). X contains the corresponding data matrix (data array)
#oversample returns two sampled arrays 
```

```
X_sample
```

```
len(X_sample)
```

```
Y_sample
```

```
len(Y_sample)
```

```
column_names = X.columns
column_names
```

```
df_X = pd.DataFrame(data = X_sample, columns = column_names ) #Making a new dataframe using the X_sample array
df_X
```

```
df_Y = pd.DataFrame(data = Y_sample, columns = ['EMB'] ) # Making a new dataframe using the Y_sample array
df_Y
```

```
new_df = pd.concat([df_X, df_Y], axis = 1) # Concaetinating the dataframe to make the final dataframe which will be used
new_df
```

<h1>**Checking for the improved ratio**

```
majority = (new_df.EMB == 1).sum() # Find the number of 1's in the output column. Just insert the column name here
print(majority)
```

```
print(majority/(new_df.shape[0] - majority))
```
