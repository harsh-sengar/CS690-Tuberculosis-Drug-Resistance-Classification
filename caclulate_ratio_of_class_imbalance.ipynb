```
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['X_trainData_column_modified_STR.csv'])) #Enter the file name here.
# Dataset is now stored in a Pandas Dataframe
```

```
df.shape[0] # Finding the total number of rows in our dataset
```

```
majority = (df.STR == 1).sum() # Find the number of 1's in the output column. Just insert the column name here
print(majority)
```

```
# Finding the ratio of the majority to minority class
from fractions import Fraction 

print(Fraction(majority/(df.shape[0] - majority)).limit_denominator())
```
